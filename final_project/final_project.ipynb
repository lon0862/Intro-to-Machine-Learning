# Data Visualization

```python
df = df.drop(['Food Description'], axis=1)
df = df.drop(['Serving'], axis=1)
df = df.drop(['Date'], axis=1)
DataFrame(df)
```

```python
df_v = df[df['type']=='vegetables']
for i in df_v.index:
  df['type'].iloc[i] = 'fruits&vegetables'

df_f = df[df['type']=='fruits']
for i in df_f.index:
  df['type'].iloc[i] = 'fruits&vegetables'
```

```python
from tabulate import tabulate
df_tmp = df.drop(['Meal'], axis=1)
type_head = ["dessert","fats&seasoning","fruits&vegetables","grains","juice","meats"]
meal_head = ['Breakfast','Lunch','Dinner','Snacks']

def show_data(food_class):
  print("\n",food_class,":")
  num = []
  for i in meal_head:
    df_meal = df[df['Meal']==i]
    num.append(df_meal[df_meal['type']==food_class].shape[0])

  num.append(df_tmp[df_tmp['type']==food_class].shape[0])
  meal_table_head = ['meal_type','Breakfast','Lunch','Dinner','Snacks','total']
  meal_table_data = [["count", "{}".format(num[0]), "{}".format(num[1]), "{}".format(num[2]), "{}".format(num[3]), "{}".format(num[4])]]
  print("\n",tabulate(meal_table_data, headers=meal_table_head, tablefmt="grid")) 

  data = [["average", "{}".format(df_tmp[df_tmp['type']==food_class].mean()[0]), "{}".format(df_tmp[df_tmp['type']==food_class].mean()[1]),
        "{}".format(df_tmp[df_tmp['type']==food_class].mean()[2]), "{}".format(df_tmp[df_tmp['type']==food_class].mean()[3]),
        "{}".format(df_tmp[df_tmp['type']==food_class].mean()[4]), "{}".format(df_tmp[df_tmp['type']==food_class].mean()[5]),],
      ["standard deviation", "{}".format(df_tmp[df_tmp['type']==food_class].std()[0]), "{}".format(df_tmp[df_tmp['type']==food_class].std()[1]),
        "{}".format(df_tmp[df_tmp['type']==food_class].std()[2]), "{}".format(df_tmp[df_tmp['type']==food_class].std()[3]),
        "{}".format(df_tmp[df_tmp['type']==food_class].std()[4]), "{}".format(df_tmp[df_tmp['type']==food_class].std()[5])]]
  head = ["feature", "Sugar","Sodium","Fat","Protein","Calories","Carbohydrates","Water Intake(fl oz)"]
  print("\n",tabulate(data, headers=head, tablefmt="grid"))

for i in type_head:
  show_data(i)
```

# Data preprocessing

```python
from sklearn.preprocessing import LabelEncoder
df_new = df.astype('category')  #transform data to number
labelencoder=LabelEncoder()
for column in df_new.columns:
    df_new[column] = labelencoder.fit_transform(df_new[column])
DataFrame(df_new)
```

```python
from sklearn.utils import shuffle #shuffle the data
df_shuff = shuffle(df_new)
```

# Train_Test_Split

```python
from sklearn.model_selection import train_test_split

for i in range(6):
  globals()['df_tmp'+str(i)] = df_shuff[df_shuff['type']==i]
  globals()['X_'+str(i)] = globals()['df_tmp'+str(i)].drop(['type'], axis=1)
  globals()['y_'+str(i)] = globals()['df_tmp'+str(i)]['type']
  globals()['X_train_'+str(i)], globals()['X_test_'+str(i)], globals()['y_train_'+str(i)], globals()['y_test_'+str(i)] = train_test_split(globals()['X_'+str(i)],globals()['y_'+str(i)],test_size=0.1,random_state=0)
  if(i==0):
    X = globals()['X_'+str(i)]
    y = globals()['y_'+str(i)]
    X_train = globals()['X_train_'+str(i)]
    y_train = globals()['y_train_'+str(i)]
    X_test = globals()['X_test_'+str(i)]
    y_test = globals()['y_test_'+str(i)]
  else:
    X = X.append(globals()['X_'+str(i)])
    y = y.append(globals()['y_'+str(i)])
    X_train = X_train.append(globals()['X_train_'+str(i)])
    y_train = y_train.append(globals()['y_train_'+str(i)])
    X_test = X_test.append(globals()['X_test_'+str(i)])
    y_test = y_test.append(globals()['y_test_'+str(i)])
```

# Model Construction

```python
from sklearn import datasets 
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
model_dt = DecisionTreeClassifier() 
model_KNN_3 = KNeighborsClassifier(n_neighbors=3) 
model_RF_100 = RandomForestClassifier(n_estimators=100) 
```

# result

```python
from sklearn.metrics import confusion_matrix
import sklearn.metrics
from sklearn.metrics import accuracy_score


def show_result(model_tmp, model_name):
  model_tmp.fit(X_train, y_train)
  y_pred_model = model_tmp.predict(X_test)

  acc = round(accuracy_score(y_test, y_pred_model)*100,2)
  rec = sklearn.metrics.recall_score(y_test,y_pred_model,average=None)
  pre = sklearn.metrics.precision_score(y_test,y_pred_model,average=None)

  print("holdout with", model_name, ":\n")
  print("\nAccuracy: {}%".format(acc))
  print("confusion matrix:")
  print(confusion_matrix(y_test, y_pred_model))

  head = ["Class Name", "Sensitivity(Recall)", "Precision"]
  data = [["dessert", "{}%".format(round(rec[0]*100,2)), "{}%".format(round(pre[0]*100,2))],
      ["fats&seasoning", "{}%".format(round(rec[1]*100,2)), "{}%".format(round(pre[1]*100,2))],
      ["fruits", "{}%".format(round(rec[2]*100,2)), "{}%".format(round(pre[2]*100,2))],
      ["grains", "{}%".format(round(rec[3]*100,2)), "{}%".format(round(pre[3]*100,2))],
      ["juice", "{}%".format(round(rec[4]*100,2)), "{}%".format(round(pre[4]*100,2))],
      ["meats", "{}%".format(round(rec[5]*100,2)), "{}%".format(round(pre[5]*100,2))]]
  print("\n",tabulate(data, headers=head, tablefmt="grid"))
```

```python
show_result(model_KNN_3, "KNN(3)")
show_result(model_dt, "Decision Tree")
show_result(model_RF_100, "Random Forest(100)")
```
